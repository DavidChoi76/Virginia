```python
import pandas as pd
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
```

```python
def hum_remover(path):
    # create df object with pandas
    df = pd.read_csv(path, sep=",", skiprows=1, skipinitialspace=True)
    t = df['Time'].values/1000000/60
    data1 = df['Temperature(F)'].values
    data2 = df['Relative Humidity(%)'].values

    fig, ax1 = plt.subplots()

    ax1.set_xlabel('Time (min)', fontsize=18)
    ax1.set_ylabel('Temperature(F)', fontsize=18)
    ax1.plot(t, data1, color='r', label='Temperature(F)', marker='o')
    ax1.tick_params(axis='y', labelsize = 15)
    ax1.legend(loc='upper center', fontsize=18)
    ax1.scatter(t, data1, s=0.1)
    
    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
    ax2.set_ylabel('Relative Humidity(%)', fontsize=18) 
    ax2.plot(t, data2, label='Relative Humidity(%)', marker='o')
    ax2.tick_params(axis='y', labelsize = 15)
    ax2.legend(loc='lower center', fontsize=18)
    ax2.scatter(t, data2, s=0.1)
       
    plt.show()
```

```python
path = "daniel.TXT"
hum_remover(path)
```

```python
path2 = "rachel.TXT"
hum_remover(path2)
```

```python
def hum_remover_hour(path):
    df = pd.read_csv(path, sep=",", skiprows=1, skipinitialspace=True)
    t = df['Time'].values/1000000/60/60/30
    data1 = df['Temperature(F)'].values
    data2 = df['Relative Humidity(%)'].values

    fig, ax1 = plt.subplots()

    ax1.set_xlabel('Time (hour)', fontsize=18)
    ax1.set_ylabel('Temperature(F)', fontsize=18)
    ax1.plot(t, data1, color='r',label='Temperature(F)')
    ax1.tick_params(axis='y', labelsize = 15)
    ax1.legend(loc='upper right', fontsize=18)
    ax1.scatter(t, data1, s=1)
    
    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
    ax2.set_ylabel('Relative Humidity(%)', fontsize=18) # we already handled the x-label with ax1
    ax2.plot(t, data2, label='Relative Humidity(%)')
    ax2.tick_params(axis='y', labelsize = 15)
    ax2.legend(loc='lower left', fontsize=18)
    ax2.scatter(t, data2, s=1)
    
    fig.tight_layout()  # otherwise the right y-label is slightly clipped

    plt.show()
```

```python
path5 = "daily_data.TXT"
hum_remover_hour(path5)
```
